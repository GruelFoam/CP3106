```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import kneighbors_graph


# Example data
np.random.seed(42)




def calculate_combined_similarity(dataframe, use_knn_affinity=True):
# Assuming `df` is your DataFrame
    embeddings = dataframe['Embeddings'].apply(lambda x: ast.literal_eval(x)).tolist()

    sic_codes = dataframe["SIC_MajorGroup"].tolist()
    naics_codes = dataframe["NAICS_SubSector"].tolist()
    cik_firm_list = dataframe["CIK"].tolist()
    #cik_mentions_list

    num_firms = len(dataframe)
    # Calculate similarities


    def create_knn_affinity_matrix(embeddings, n_neighbors=10):
        knn_graph = kneighbors_graph(embeddings, n_neighbors=n_neighbors, mode='connectivity', include_self=True)
        knn_affinity = 0.5 * (knn_graph + knn_graph.T)  # Make it symmetric
        return knn_affinity.toarray()
    # Assuming `df` is your DataFrame


    n_neighbors = 10
    knn_affinity = create_knn_affinity_matrix(embeddings, n_neighbors=n_neighbors)

    embedding_similarity = cosine_similarity(embeddings)
    if(use_knn_affinity == True):
        embedding_similarity = knn_affinity


    combined_similarity = embedding_similarity #just return embedding similarity thats all

    return combined_similarity
```

```python
def compute_similarity_matrix_for_all_years(dataframe, year_list_range,use_knn_affinity=True):
    combined_similarity_dict = {}

    for year in year_list_range:
        focal_df = dataframe[dataframe.Year == year]
        combined_similarity = calculate_combined_similarity(focal_df,  use_knn_affinity )
        combined_similarity_dict[year] = combined_similarity

    return combined_similarity_dict
```

```python
def intra_industry_correlations(df, classification_column):
    intra_matrix = []
    industry_groups = df.groupby(classification_column)
    intra_correlations = {}
    for industry, group in industry_groups:
        pivot_table = group.pivot(index='Date', columns='tic', values='Return')
        corr_matrix = pivot_table.corr()
        # Take the upper triangle of the correlation matrix without the diagonal
        intra_corr = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)).stack().mean()
        intra_correlations[industry] = intra_corr
    return intra_correlations
```

```python
def fetch_stock_data(df_full_set, year, mode):
    # Filter data for the given year
    df_processed = df_full_set[df_full_set.Year == year]

    # Get unique tickers
    tickers = df_processed['tic'].unique().tolist()

    # Define the period based on the mode
    if mode == 'in-sample':
        # In-sample mode: start and end within the same year
        start_date = f'{year}-01-01'
        end_date = f'{year}-12-31'
    elif mode == 'forward':
        # Forward mode: next two years
        start_date = f'{year+1}-01-01'
        end_date = f'{year+2}-12-31'
    elif mode == 'backtest':
        # Backtest mode: last 10 years, up until year-1
        start_date = f'{year-10}-01-01'
        end_date = f'{year-1}-12-31'
    else:
        raise ValueError("Mode must be 'in-sample', 'forward', or 'backtest'")

    # Fetch historical stock data for all tickers
    stock_data = yf.download(tickers, start=start_date, end=end_date)

    # Get adjusted close prices
    close_prices = stock_data['Adj Close']

    # Drop columns where all values are NaN
    close_prices = close_prices.dropna(axis=1, how='all')

    # Calculate daily returns
    returns = close_prices.pct_change()

    # Melt the returns DataFrame to long format for merging
    returns_long = returns.reset_index().melt(id_vars='Date', var_name='tic', value_name='Return')



# Merge the returns with the original dataframe



    return returns_long
```

```python
from hyperopt import fmin, tpe, hp, Trials, STATUS_OK
import numpy as np

def create_knn_affinity_matrix(embeddings, n_neighbors=10):
    knn_graph = kneighbors_graph(embeddings, n_neighbors=n_neighbors, mode='connectivity', include_self=True)
    knn_affinity = 0.5 * (knn_graph + knn_graph.T)  # Make it symmetric
    return knn_affinity.toarray()
    # Assuming `df` is your DataFrame



# Define the objective function for Hyperopt
def objective(params, n_clusters, focal_returns, focal_df):
    weight_embeddings = params['weight_embeddings']
    weight_sic = params['weight_sic']
    weight_naics = params['weight_naics']
    weight_mentions = params['weight_mentions']


    n_neighbors = 10
    knn_affinity = create_knn_affinity_matrix(embeddings, n_neighbors=n_neighbors)

    # Ensure weights sum to 1
    total_weight = weight_embeddings + weight_sic + weight_naics + weight_mentions
    weight_embeddings /= total_weight
    weight_sic /= total_weight
    weight_naics /= total_weight
    weight_mentions /= total_weight

    # Assuming combined similarity matrix calculation and df_processed are available
    combined_similarity = calculate_combined_similarity(focal_df, weight_embeddings,weight_sic, weight_naics,weight_mentions, use_knn_affinity = True)



    # Perform Spectral Clustering with the passed n_clusters
    spectral = SpectralClustering(n_clusters=n_clusters, affinity='precomputed', random_state=42)
    focal_df["emb_a"] = spectral.fit_predict(combined_similarity)

    # Merge with returns data and compute intra-industry correlations
    df_returns = pd.merge(focal_df, focal_returns, on='tic', how='inner')
    intra_corrs = intra_industry_correlations(df_returns, "emb_a")
    avg_intra_corr = np.nanmean(list(intra_corrs.values()))

    # Return negative of avg_intra_corr because Hyperopt minimizes the objective
    return {'loss': -avg_intra_corr, 'status': STATUS_OK}

# Define the search space for the weights
space = {
    'weight_embeddings': hp.uniform('weight_embeddings', 0, 1),
    'weight_sic': hp.uniform('weight_sic', 0, 1),
    'weight_naics': hp.uniform('weight_naics', 0, 1),
    'weight_mentions': hp.uniform('weight_mentions', 0, 1)
}

# Function to run optimization for a given number of clusters
def run_experiment_for_n_clusters(n_clusters, focal_returns, focal_df):
    trials = Trials()
    best_params = fmin(
        fn=lambda params: objective(params, n_clusters, focal_returns, focal_df),
        space=space,
        algo=tpe.suggest,
        max_evals=25,
        trials=trials
    )
    print(f"Best parameters for {n_clusters} clusters: ", best_params)
    return best_params

#df_in_focus = df_full_set[df_full_set.Year == 2021]
#returns_long = fetch_stock_data(df_full_set, 2021, "in-sample")
# Run separate optimizations for 10 and 100 clusters
#best_params_10 = run_experiment_for_n_clusters(10, returns_long,df_in_focus )
#best_params_100 = run_experiment_for_n_clusters(100, returns_long,df_in_focus)
```
