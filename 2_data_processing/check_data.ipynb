### This piece of code help check the correctness of [final_df_cleaned]

```python
import pandas as pd
```

```python
file_path = '../item1_and_financial_variable/multi-view/final_df_cleaned.csv'

data = pd.read_csv(file_path)

print(data.head())
```

```python
selected = data[['Unnamed: 0', 'conm', 'tic', 'SP_BUSINESS_DESCRIPTION', 'SP_LONG_BUSINESS_DESCRIPTION', 'SP_COMPANY_NAME', 'YH_BUS_DESC', 'SA_BUS_DESC']]
```

```python
type(selected)
```

```python
selected['SP_BUSINESS_DESCRIPTION'] = selected['SP_BUSINESS_DESCRIPTION'].astype(str)

selected['SP_LONG_BUSINESS_DESCRIPTION'] = selected['SP_LONG_BUSINESS_DESCRIPTION'].astype(str)

selected['YH_BUS_DESC'] = selected['YH_BUS_DESC'].astype(str)

selected['SA_BUS_DESC'] = selected['SA_BUS_DESC'].astype(str)

selected['SP_COMPANY_NAME'] = selected['SP_COMPANY_NAME'].astype(str)
```

```python
selected.columns
```

```python
def basic_filter(row):
    prefix = row['SP_COMPANY_NAME']
    if prefix == "nan":
        return row
    
    if row['SP_BUSINESS_DESCRIPTION'] != "nan" and not row['SP_BUSINESS_DESCRIPTION'].startswith(prefix):
        return row
    if row['SP_LONG_BUSINESS_DESCRIPTION'] != "nan" and not row['SP_LONG_BUSINESS_DESCRIPTION'].startswith(prefix):
        return row
    if row['YH_BUS_DESC'] != "nan" and not row['YH_BUS_DESC'].startswith(prefix):
        return row
    if row['SA_BUS_DESC'] != "nan" and not row['SA_BUS_DESC'].startswith(prefix):
        return row
    
suspicious = selected.apply(basic_filter, axis=1).dropna()
```

```python
suspicious
```

```python
suspicious['SP_BUSINESS_DESCRIPTION'] = suspicious['SP_BUSINESS_DESCRIPTION'].apply(lambda x: x[:150])

suspicious['SP_LONG_BUSINESS_DESCRIPTION'] = suspicious['SP_LONG_BUSINESS_DESCRIPTION'].apply(lambda x: x[:150])

suspicious['YH_BUS_DESC'] = suspicious['YH_BUS_DESC'].apply(lambda x: x[:150])

suspicious['SA_BUS_DESC'] = suspicious['SA_BUS_DESC'].apply(lambda x: x[:150])

```
